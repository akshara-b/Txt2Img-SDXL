```python
from diffusers import StableDiffusionXLPipeline
import torch

# Load the base SDXL model
pipe = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,  # Saves memory
    use_safetensors=True        # Faster loading
)

# Move to GPU (Colab provides one)
pipe = pipe.to("cuda")
```

```python
prompt = "a dog and a cat watching tv"
image = pipe(prompt).images[0]  # Generate the image
image.save("generated_image.png")  # Save it
```

```python
import gradio as gr

def generate_image(prompt):
    image = pipe(prompt).images[0]
    return image

# Create a simple interface
interface = gr.Interface(
    fn=generate_image,
    inputs="text",
    outputs="image",
    title="SDXL Text-to-Image Generator"
)

interface.launch(share=True)  # Share a public link
```
